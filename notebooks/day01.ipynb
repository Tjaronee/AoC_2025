```python
# Load input data for day 1
data = get_input(1)
print(f"Loaded {len(data)} lines of input")
print(f"First line: {data[0] if data else 'No data'}")
```

## Part 1

```python
# create library from data where the first value is either L or R 
# and the second value is the number of steps to take in that direction
directions = [(line[0], int(line[1:])) for line in data]
print(directions[:5])  # Print first 5 directions to verify
```

```python
mod_value = 100
initial_pos = 50
```

```python
position = initial_pos
count_when_zero = 0

for turn, steps in directions:
    if turn == 'L':
        position == steps
    elif turn == 'R':
        position += steps
    position %= mod_value

    if position == 0:
        count_when_zero += 1

print(f"count_when_zero: {count_when_zero}")
```

## Part 2

```python
position = initial_pos

# count when zero is passed, ie when position wraps around beyond zero or mod_value
count_when_zero_is_passed = 0
for turn, steps in directions:
    if turn == 'L':
        sign = -1
    elif turn == 'R':
        sign = 1
    for _ in range(steps):
        position += sign
        position %= mod_value
        if position == 0 or position == mod_value:
            count_when_zero_is_passed += 1

print(f"count_when_zero_is_passed: {count_when_zero_is_passed}")
```
