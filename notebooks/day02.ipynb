```python
# Setup: Add project root to path and import utilities
import sys; sys.path.insert(0, '..')
from utils import get_input
```

```python
# Load input data for day 2
data = get_input(2)

print(f"Loaded {len(data)} lines of input")
print(f"First line: {data[0] if data else 'No data'}")
```

```python
# split data
data2 = []
for line in data:
    ranges = line.split(',')
    for range_str in ranges:
        parts = range_str.split('-')
        data2.append((int(parts[0]), int(parts[1])))
```

```python
def multiple_repeats(str_value, n):
   # find if str_value[0:n] repeats to make str_value
   # example: str_value='123123', n=3 gives str_value[0:3]='123' and str_value[3:6]='123' -> True
    length = len(str_value)
    if length % n != 0:
        return False
    segment = str_value[0:n]
    for i in range(0, length, n):
        if str_value[i:i+n] != segment:
            return False
    return True
    
```

```python
output = 0
for line in data2:
    # print(f"Processing range: {line[0]} to {line[1]}")
    for value in range(line[0], line[1]+1):
        str_value = str(value)

        length_is_even = len(str_value) % 2 == 0
        midway_lenghth = len(str_value) // 2

        if length_is_even and multiple_repeats(str_value, midway_lenghth):
            output += value
print(f"Output: {output}")
```

```python
output = 0
for line in data2:
    for value in range(line[0], line[1]+1):
        str_value = str(value)

        for n in range(1, len(str_value)//2+1):
            length_is_divisible = len(str_value) % n == 0
            if length_is_divisible and multiple_repeats(str_value, n):
                output += value 
                break
print(f"Output: {output}")
```

```python
# test cases for multiple_repeats
assert multiple_repeats('123123', 2) == False
assert multiple_repeats('123123', 3) == True
assert multiple_repeats('123124', 3) == False
assert multiple_repeats('11112222', 3) == False
assert multiple_repeats('1010', 2) == True
```
