```python
# Setup: Add project root to path and import utilities
import sys; sys.path.insert(0, '..')
from utils import get_input

# Load input data for day 4
data = get_input(4)

print(f"Loaded {len(data)} lines of input")
print(f"First line: {data[0] if data else 'No data'}")
```

```python
# create a grid from the input data
import numpy as np
grid = np.array([[char for char in line] for line in data])
print("Grid shape:", grid.shape)
print("Grid sample:\n", grid[:5, :5])
```

```python
def neighboorhood(x, y, grid):
    """Get the values of the 8 neighboring cells around (x, y) in the grid, and remember coordinates"""
    neighbors = {}  # use a dict instance, not the dict type
    for dx in [-1, 0, 1]:
        for dy in [-1, 0, 1]:
            if dx == 0 and dy == 0:
                continue
            nx, ny = x + dx, y + dy
            if 0 <= nx < grid.shape[0] and 0 <= ny < grid.shape[1]:
                # add neighbor with its coordinates
                neighbors[(nx, ny)] = grid[nx, ny]
    return neighbors
```

```python
def nr_of_papers_in_neighboorhood(x, y, grid):
    neighbors = neighboorhood(x, y, grid)
    # count how many neighbors have the value '@'
    return sum(1 for value in neighbors.values() if value == '@')
```

```python
def how_many_papers_can_be_accessed(grid):
    count = 0
    for x,row in enumerate(grid):
        for y, cell in enumerate(row):
            if cell == '@':  
                if nr_of_papers_in_neighboorhood(x, y, grid) < 4:
                    count += 1
    return count

print("Number of papers that can be accessed:", how_many_papers_can_be_accessed(grid))
```

```python
def remove_and_count_papers(grid):
    removed_count = 0
    new_grid = grid.copy()
    for x,row in enumerate(grid):
        for y, cell in enumerate(row):
            if cell == '@':
                if nr_of_papers_in_neighboorhood(x, y, grid) < 4:
                    new_grid[x, y] = '.'  # remove paper
                    removed_count += 1
    return new_grid, removed_count
```

```python
total_removed = 0
while True:
    grid, removed = remove_and_count_papers(grid)
    total_removed += removed
    if removed == 0:
        break

print(f"total removed: {total_removed}")  
```
