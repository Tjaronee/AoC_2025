```python
# Setup: Add project root to path and import utilities
import sys; sys.path.insert(0, '..')
from utils import get_input

# Load input data for day 6
data = get_input(6)

print(f"Loaded {len(data)} lines of input")
for line in data:
    print(line)
```

```python
# Parse data: split by spaces, convert to integers where possible
clean_data = []
for line in data:
    values = line.split(' ')
    # Remove empty strings and convert digits to integers
    values = [int(v) if v.isdigit() else v for v in values if v]
    clean_data.append(values)

# Display parsed data
for line in clean_data:
    print(line)
```

```python
# Part 1: Calculate sum of results using operations in last row
nr_of_equations = len(clean_data[0])
operations = clean_data[-1]

sum_of_solutions = 0
for i in range(nr_of_equations):
    values = [line[i] for line in clean_data[:-1]]
    operation = operations[i]
    
    if operation == '+':
        result = sum(values)
    elif operation == '*':
        result = 1
        for v in values:
            result *= v
    else:
        raise ValueError(f"Unknown operation: {operation}")
    
    sum_of_solutions += result

print(f"Sum of solutions (Part 1): {sum_of_solutions}")
```

```python
funky_clean_data = []
nr_of_raw_columns = len(data[0])

# last line needs one more bitsize due to end of data
equation_bit_size = 1
collected_values = []
for i in reversed(range(nr_of_raw_columns)):
    operation = data[-1][i]
    empty_operation = operation == ' '

    # value is all numbers in the column top to bottom, skipping empty strings
    value = [line[i] for line in data[:-1] if line[i] != ' ']

    # convert to number
    value = ''.join(value)
    # if value is a number append
    if value.isdigit():
        collected_values.append(int(value))

    if empty_operation:
        equation_bit_size += 1
        continue
    else:
        # append collected values to funky_clean_data
        funky_clean_data.append((list(reversed(collected_values)),operation))
        # print(f"Collected equation with bit size {equation_bit_size}: {list(reversed(collected_values))}")
    
    # print(f"Collected values: {collected_values}")
    # reset bit size for next equation
    equation_bit_size = 0
    collected_values = []

for line in clean_data:
    print(line)
print(list(reversed(funky_clean_data)))
```

```python
# Part 2: Calculate sum of results using column-parsed equations
sum_of_solutions_part2 = 0
for values, operation in funky_clean_data:
    if operation == '+':
        result = sum(values)
    elif operation == '*':
        result = 1
        for v in values:
            result *= v
    else:
        raise ValueError(f"Unknown operation: {operation}")
    
    sum_of_solutions_part2 += result

print(f"Sum of solutions (Part 2): {sum_of_solutions_part2}")
```
